```python
import pandas as pd

# URL del conjunto de datos
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/Station_Entries_Daily_Totals.csv'

# Cargar el conjunto de datos en un DataFrame de Pandas
data = pd.read_csv(url)

# Mostrar las primeras filas del DataFrame para verificar que se cargó correctamente
print(data.head())
```

```python
# Mostrar la forma del DataFrame (número de filas y columnas)
print("Shape of the DataFrame:")
print(data.shape)

# Mostrar las primeras filas del DataFrame
print("\nFirst few rows of the DataFrame:")
print(data.head())

# Mostrar las últimas filas del DataFrame
print("\nLast few rows of the DataFrame:")
print(data.tail())

# Mostrar los tipos de datos de cada columna
print("\nData types of each column:")
print(data.dtypes)

# Mostrar el tipo del objeto DataFrame
print("\nType of the DataFrame:")
print(type(data))
```

```python
# Mostrar estadísticas descriptivas del DataFrame
print("Descriptive statistics of the DataFrame:")
print(data.describe())
```

```python
# Encontrar los valores mínimos en cada columna
min_values = data.min()

# Encontrar los valores máximos en cada columna
max_values = data.max()

# Mostrar los valores mínimos
print("Minimum values in each column:")
print(min_values)

# Mostrar los valores máximos
print("\nMaximum values in each column:")
print(max_values)
```

```python
import pandas as pd

# URL del conjunto de datos
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/Station_Entries_Daily_Totals.csv'

# Cargar el conjunto de datos en un DataFrame de pandas
df = pd.read_csv(url)

# Imprimir los nombres de las columnas
print("Columnas del DataFrame:", df.columns)

# Convertir la columna de fechas a formato datetime
df['date'] = pd.to_datetime(df['date'])  # Aquí cambiamos 'Date' a 'date'

# Establecer la columna de fechas como índice del DataFrame
df.set_index('date', inplace=True)  # Aquí también cambiamos 'Date' a 'date'

# Visualizar las primeras filas del DataFrame después de la modificación
print(df.head())
```

```python
# Ordenar el DataFrame por la columna de fechas
df_sorted = df.sort_values(by='date')

# Visualizar las primeras filas del DataFrame ordenado
print(df_sorted.head())
```

```python
# Filtrar el DataFrame para la estación 'Morgan-Lake'
Morgan_Lake = df[df['stationname'] == 'Morgan-Lake']

# Visualizar las primeras filas del nuevo DataFrame
print(Morgan_Lake.head())
```

```python
# Filtrar el DataFrame para la estación 'Addison-Brown'
Addison_Brown = df[df['stationname'] == 'Addison-Brown']

# Visualizar las primeras filas del nuevo DataFrame
print(Addison_Brown.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer el tamaño de la figura
sns.set(rc={'figure.figsize':(11, 4)})

# Trama de la columna 'rides'
plt.plot(Morgan_Lake.index, Morgan_Lake['rides'], linewidth=0.5)

# Configuración de la trama
plt.title('Número de viajes en la estación Morgan-Lake')
plt.xlabel('Fecha')
plt.ylabel('Número de viajes')

# Mostrar la trama
plt.show()
```
